
# Top 5 Data Visualization Libraries in Python in 2023

## Introduction

Data visualization is an integral part of data analysis, providing insights through visual representation. Python, a leader in the data science world, offers a plethora of libraries for this purpose. In 2023, some libraries have stood out for their ease of use, comprehensive features, and robust community support. Let's explore the top 5 data visualization libraries in Python.

### 1. Matplotlib

Overview: Matplotlib is the grandfather of Python data visualization libraries, offering immense control over every element of a plot.

Key Features: Its versatility allows creation of simple line plots to complex 3D diagrams.

#### Pros:

* High level of customization for creating precise plots.
* Supports a wide range of formats and backends.
* Strong community support and extensive documentation.
#### Cons:

* Steeper learning curve for advanced features.
* Less suited for interactive visualizations.
* Outdated default visual style (though customizable).



```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

### 2. Seaborn

Overview: Seaborn extends Matplotlib with more visually appealing and complex plotting options.

Key Features: It's particularly well-suited for statistical data visualizations.

#### Pros:

* Simplifies creation of complex statistical plots.
* Integrates well with Pandas data structures.
* Provides beautiful default themes and color palettes.
#### Cons:

* Limited to statistical visualizations.
* Less flexibility compared to Matplotlib for non-standard plots.
* Dependent on Matplotlib for low-level functions.

```python

import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': range(1, 11), 'y': range(10, 110, 10)})
sns.heatmap(data.corr())
```

### 3. Plotly

Overview: Plotly specializes in creating interactive and aesthetically pleasing visualizations that can be viewed in a web browser.
Key Features: It excels in interactive and 3D plots.

#### Pros:

* Highly interactive and dynamic visualizations.
* Web integration capabilities (Plotly Dash).
* Supports 3D plots and complex chart types.
#### Cons:

* Larger output file size, which can be slower to load.
* Can be more resource-intensive for complex visualizations.
* Some features require an internet connection.



```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "vscode"
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length', color="species")

# all three of these worked
fig.show()
```

### 4. Altair

Overview: Altair, a declarative statistical visualization library, is recognized for its simplicity and efficiency.

Key Features: It translates Python code into a JSON format which is then rendered by Vega-Lite in the browser.


#### Pros:

* Simple and intuitive syntax for building plots.
* Automatic data type inference, making plotting easier.
* Effective for exploratory data analysis.
#### cons:

* Performance can degrade with large datasets.
* Less community support and resources compared to others.
* Limited customization for highly specific visualization needs.


```python
import altair as alt
import pandas as pd

# Example data
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [3, 1, 4, 6, 5]
})

# Create a scatter plot
scatter_chart = alt.Chart(data).mark_circle(size=60).encode(
    x='x',
    y='y',
    tooltip=['x', 'y']  # This adds a tooltip on hover
)

# To display in a Jupyter notebook, simply call the chart object
scatter_chart
```

### 5. Bokeh

Overview: Bokeh is ideal for those who need interactive plots that can be easily embedded in web applications.

Key Features: It allows for streaming data and real-time updates, which is a boon for dynamic data visualizations.

#### Pros:

* Powerful interactive features, ideal for web applications.
* Supports streaming and real-time data.
* Good integration with other web technologies.
#### Cons:

* Steeper learning curve for beginners.
* Performance issues with very large datasets.
* Limited variety of charts compared to some other libraries.


```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

# Activate inline plotting for Jupyter notebooks
output_notebook()

# Sample data
data = pd.DataFrame({
    'x_values': range(10),
    'y_values': [5, 7, 2, 2, 4, 6, 8, 6, 2, 5]
})

# Convert DataFrame to a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot with a title and axis labels
p = figure(title="Simple line example", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line('x_values', 'y_values', source=source, legend_label="Temp.", line_width=2)

# Show the results
show(p)
```
